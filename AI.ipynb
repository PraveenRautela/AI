```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hypothyroid.csv')
```

```python
df.head()
```

```python
columns = ['target', 'age', 'sex', 'on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_medication',
           'thyroid_surgery', 'query_hypothyroid', 'query_hyperthyroid', 'pregnant', 'sick', 'tumor',
           'lithium', 'goitre', 'TSH_measured', 'TSH', 'T3_measured', 'T3', 'TT4_measured', 'TT4',
           'T4U_measured', 'T4U', 'FTI_measured', 'FTI', 'TBG_measured', 'TBG']
```

```python
df.columns = columns
```

```python
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
df['target'] = df['target'].apply(lambda x: 1 if x == 'hypothyroid' else 0)
```

```python
df['sex'] = df['sex'].replace({'M': 1, 'F': 0})
```

```python
binary_columns = ['on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_medication', 'thyroid_surgery',
                  'query_hypothyroid', 'query_hyperthyroid', 'pregnant', 'sick', 'tumor', 'lithium', 'goitre',
                  'TSH_measured', 'T3_measured', 'TT4_measured', 'T4U_measured', 'FTI_measured', 'TBG_measured']
```

```python
df[binary_columns] = df[binary_columns].replace({'f': 0, 't': 1, 'n': 0, 'y': 1})
```

```python
df.head()
```

```python
features = ['age', 'sex', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'goitre', 'thyroid_surgery', 'query_hypothyroid', 'query_hyperthyroid']
```

```python
df['T3'] = df['T3'].replace('?', np.nan)
df['TT4'] = df['TT4'].replace('?', np.nan)
df['TSH'] = df['TSH'].replace('?', np.nan)
```

```python
df['T3'] = df['T3'].astype(float)
df['TT4'] = df['TT4'].astype(float)
df['TSH'] = df['TSH'].astype(float)
```

```python
df1 = df[features]
x = pd.DataFrame(df1)
```

```python
df2 = df['target']
y = pd.DataFrame(df2)
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
```

```python
model = BayesianNetwork([('T3', 'target'), ('TT4', 'target'), ('TSH', 'target'), ('goitre', 'target')])
```

```python
df_combined = pd.concat([X_train[features], y_train['target']], axis=1)
```

```python
model.fit(df_combined, estimator=MaximumLikelihoodEstimator)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test[['TT4','T3','TSH','goitre']])

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
